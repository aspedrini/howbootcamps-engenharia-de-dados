```python
#cria uma 'coleção' e conta quantas vezes ela apareceu
counter = collections.Counter(comb)
counter
```

```python
resultado = pd.DataFrame(counter.items(), columns = ['Combinacao', 'Frequencia'])
resultado
```

```python
resultado['p_freq'] = resultado['Frequencia']/(resultado['Frequencia'].sum())
resultado
```

```python
resultado = resultado.sort_values(by='p_freq', ascending= False)
resultado
```
