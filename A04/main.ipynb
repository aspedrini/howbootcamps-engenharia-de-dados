```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine(
    'postgresql+psycopg2://root:root@localhost/test_db'
)
```

```python
sql = '''
select * from vw_artist
'''
```

```python
df = pd.read_sql_query(sql, engine)
df
```

```python
#another way:
df_song = pd.read_sql_query('select * from tb_web', engine)
df_song
```

```python
# execute a query inside our postgresql
sql2 = '''
insert into public.tb_artist (select
	t1."date",
	t1."rank",
	t1.artist, t1.song
from
	public."Billboard" t1
where t1.artist like 'Metallica'
order by
	t1.artist,
	t1."date")
'''
```

```python
engine.execute(sql2)
```
